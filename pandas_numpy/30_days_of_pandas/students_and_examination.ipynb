```python
import pandas as pd

# Sample data for Students table
students_data = {
    'student_id': [1, 2, 13, 6],
    'student_name': ['Alice', 'Bob', 'John', 'Alex']
}
students_df = pd.DataFrame(students_data)

# Sample data for Subjects table
subjects_data = {
    'subject_name': ['Math', 'Physics', 'Programming']
}
subjects_df = pd.DataFrame(subjects_data)

# Sample data for Examinations table
examinations_data = {
    'student_id': [1, 1, 1, 2, 1, 1, 13, 13, 13, 2, 1],
    'subject_name': ['Math', 'Physics', 'Programming', 'Programming', 'Physics', 'Math', 'Math', 'Programming', 'Physics', 'Math', 'Math']
}
examinations_df = pd.DataFrame(examinations_data)

# Create the Cartesian join (cross join) between Students and Subjects
students_df['key'] = 1
subjects_df['key'] = 1
cartesian_join = pd.merge(students_df, subjects_df, on='key').drop('key', axis=1)

# Merge the result with the Examinations table
students_and_exams = pd.merge(cartesian_join, examinations_df, on=['student_id', 'subject_name'], how='left')

# Fill NaN values in 'subject_name' with 0 to count attended exams
students_and_exams['attended_exams'] = students_and_exams['subject_name'].notna().astype(int)

# Group by and count the attended exams
results = students_and_exams.groupby(['student_id', 'student_name', 'subject_name']).size().reset_index(name='attended_exams')

print(results)
```
